## Wilcoxon Rank Sum Tests

```python
filtered_csv_data2 = pd.read_csv('200gene_dge_data.csv')
filtered_df = pd.DataFrame(filtered_csv_data2)
```

```python
SNPpairs = list()
for i in range(len(influential_SNPs)):
        for j in range(i+1, len(influential_SNPs)):
            SNPpairs = SNPpairs + [(influential_SNPs[i], influential_SNPs[j])]

SNPpairs
```

```python
drop_genes = list()

for i in SNPpairs:
    a = i[0]
    b = i[1]
    c = ranksums(x=filtered_df[a], y=filtered_df[b])
    
    if c[1] < .05:
        drop_genes.append(c)
    
np.unique(drop_genes)
```
