```python
pd.set_option('display.max_columns', 500)
```

```python
all_data = pd.read_csv('phenotypes_202304020100.csv', sep=';')
all_data_df = pd.DataFrame(all_data)
print(all_data_df)
```

```python
cb_df = all_data_df[all_data_df['Colour Blindness'] != '-']
print(cb_df)
```

```python
cb_responses = cb_df['Colour Blindness']
print(cb_responses)

gen_cb = 0
rg_cb = 0
yb_cb = 0
not_cb = 0

tokens_lower = []

for i in cb_responses:
  
    #tokens = nltk.word_tokenize(i) #breaks up each response into words
    tokens = str(i)
    tokens = nltk.word_tokenize(tokens) #breaks up each response into words
    tokens_lc = [item.lower() for item in tokens]
    tokens_lower.append(tokens_lc)


for i in tokens_lower:
    if 'red' in i:
        gen_cb = gen_cb + 1
        rg_cb = rg_cb + 1
  
    if 'blue' in i:
        gen_cb = gen_cb + 1
        yb_cb = yb_cb + 1

    if 'false' in i:
        not_cb = not_cb + 1

    if 'no' in i:
        not_cb = not_cb + 1

    if 'true' in i:
        gen_cb = gen_cb + 1

    if 'yes' in i:
        gen_cb = gen_cb + 1
  
    if 'not' in i:
        not_cb = not_cb + 1
```
